Summary:

Default axis value: 0 (vertical concatenation)

axis=0: Stack rows → increases row count

axis=1: Stack columns → increases column count

```python
# Create a function to analyze team performance
def analyze_team_performance(df):
    # Get all unique teams
    all_teams = pd.unique(df[['HomeTeam', 'AwayTeam']].values.ravel('K'))

    team_data = []

    for team in all_teams:
        # Home matches
        home_matches = df[df['HomeTeam'] == team]
        # Away matches
        away_matches = df[df['AwayTeam'] == team]

        # Home statistics
        home_goals_for = home_matches['FTHG'].sum()
        home_goals_against = home_matches['FTAG'].sum()
        home_wins = len(home_matches[home_matches['FTR'] == 'H'])
        home_draws = len(home_matches[home_matches['FTR'] == 'D'])
        home_losses = len(home_matches[home_matches['FTR'] == 'A'])

        # Away statistics
        away_goals_for = away_matches['FTAG'].sum()
        away_goals_against = away_matches['FTHG'].sum()
        away_wins = len(away_matches[away_matches['FTR'] == 'A'])
        away_draws = len(away_matches[away_matches['FTR'] == 'D'])
        away_losses = len(away_matches[away_matches['FTR'] == 'H'])

        # Totals
        total_matches = len(home_matches) + len(away_matches)
        total_goals_for = home_goals_for + away_goals_for
        total_goals_against = home_goals_against + away_goals_against
        total_wins = home_wins + away_wins
        total_draws = home_draws + away_draws
        total_losses = home_losses + away_losses

        team_data.append({
            'Team': team,
            'Matches': total_matches,
            'Wins': total_wins,
            'Draws': total_draws,
            'Losses': total_losses,
            'Goals_For': total_goals_for,
            'Goals_Against': total_goals_against,
            'Goal_Difference': total_goals_for - total_goals_against,
            'Points': total_wins * 3 + total_draws,
            'Home_Goals': home_goals_for,
            'Away_Goals': away_goals_for
        })

    return pd.DataFrame(team_data)

# Create team performance dataframe
team_performance = analyze_team_performance(df_combined)
team_performance_sorted = team_performance.sort_values('Points', ascending=False)

print("=== TEAM PERFORMANCE RANKING ===")
print(team_performance_sorted[['Team', 'Points', 'Goals_For', 'Goals_Against', 'Goal_Difference', 'Wins', 'Draws', 'Losses']].head(10))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime

print("=" * 80)
print("🌟 FREESTYLE EXPLORATION & ADDITIONAL INSIGHTS")
print("=" * 80)

# First, let's make sure we have the data loaded
print("\n" + "🔍 " + "=" * 50)
print("🔍 DATA CHECK & PREPARATION")
print("🔍 " + "=" * 50)

# Check if df1819 exists, if not load it
try:
    # Test if df1819 exists
    _ = df1819.shape
    print("✅ df1819 is already loaded")
    print(f"   Shape: {df1819.shape}")
except NameError:
    print("❌ df1819 not found. Please load your data first.")
    print("   Example: df1819 = pd.read_excel('your_file.xlsx')")
    # Exit gracefully if data isn't loaded
    print("\n📍 Please load your data and run this code again.")
    exit()

# Also ensure df_tot exists for team analysis
try:
    _ = df_tot.shape
    print("✅ df_tot is available for team analysis")
except NameError:
    print("⚠️  df_tot not found. Creating it now...")
    # Create df_tot if it doesn't exist
    groupedH = df1819.groupby('HomeTeam')['FTHG'].sum()
    groupedA = df1819.groupby('AwayTeam')['FTAG'].sum()

    df_groupedA = pd.DataFrame(groupedA)
    df_groupedH = pd.DataFrame(groupedH)
    df_groupedA['team'] = groupedA.index
    df_groupedH['team'] = groupedH.index
    df_groupedA.columns = ['away_goals', 'team']
    df_groupedH.columns = ['home_goals', 'team']

    df_tot = pd.merge(df_groupedH, df_groupedA, on='team', how='outer').fillna(0)
    df_tot['total'] = df_tot['home_goals'] + df_tot['away_goals']
    print("✅ df_tot created successfully")

# 1. Goal Scoring Patterns Analysis
print("\n" + "⚽ " + "=" * 50)
print("⚽ GOAL SCORING PATTERNS & TRENDS")
print("⚽ " + "=" * 50)

# Calculate goals per match distribution
print("📍 Match Goal Distribution Analysis:")
total_goals_per_match = df1819['FTHG'] + df1819['FTAG']
goal_ranges = {
    '0-1 goals': ((total_goals_per_match <= 1).sum()),
    '2-3 goals': ((total_goals_per_match >= 2) & (total_goals_per_match <= 3)).sum(),
    '4-5 goals': ((total_goals_per_match >= 4) & (total_goals_per_match <= 5)).sum(),
    '6+ goals': ((total_goals_per_match >= 6)).sum()
}

for range_name, count in goal_ranges.items():
    percentage = (count / len(df1819)) * 100
    print(f"   • {range_name}: {count} matches ({percentage:.1f}%)")

# 2. Team Performance Clustering
print("\n" + "📊 " + "=" * 50)
print("📊 TEAM PERFORMANCE CLUSTERING")
print("📊 " + "=" * 50)

# Create performance metrics
team_stats = pd.DataFrame({
    'Total_Goals': df_tot['total'],
    'Home_Goals': df_tot['home_goals'],
    'Away_Goals': df_tot['away_goals'],
    'Home_Ratio': (df_tot['home_goals'] / df_tot['total'] * 100).round(1)
}, index=df_tot['team'])

# Categorize teams
def categorize_team(row):
    if row['Total_Goals'] >= team_stats['Total_Goals'].quantile(0.75):
        return 'High Scorer'
    elif row['Total_Goals'] <= team_stats['Total_Goals'].quantile(0.25):
        return 'Low Scorer'
    else:
        return 'Medium Scorer'

team_stats['Category'] = team_stats.apply(categorize_team, axis=1)

print("📍 Team Performance Categories:")
for category in ['High Scorer', 'Medium Scorer', 'Low Scorer']:
    teams = team_stats[team_stats['Category'] == category].index.tolist()
    print(f"   • {category}: {', '.join(teams)}")

# 3. Monthly Performance Trends
print("\n" + "📅 " + "=" * 50)
print("📅 MONTHLY PERFORMANCE TRENDS")
print("📅 " + "=" * 50)

# Convert date and analyze monthly trends
try:
    df1819['Date'] = pd.to_datetime(df1819['Date'])
    df1819['Month'] = df1819['Date'].dt.month_name()

    monthly_goals = df1819.groupby('Month').agg({
        'FTHG': 'sum',
        'FTAG': 'sum'
    })
    monthly_goals['Total'] = monthly_goals['FTHG'] + monthly_goals['FTAG']

    print("📍 Monthly Goal Distribution:")
    for month, row in monthly_goals.iterrows():
        print(f"   • {month}: {int(row['Total'])} total goals")

except Exception as e:
    print(f"   ⚠️ Date analysis not available: {e}")

# 4. Head-to-Head Rivalries
print("\n" + "🔥 " + "=" * 50)
print("🔥 RIVALRY MATCH ANALYSIS")
print("🔥 " + "=" * 50)

# Find most common matchups
matchup_counts = df1819.groupby(['HomeTeam', 'AwayTeam']).size().sort_values(ascending=False)
top_matchups = matchup_counts.head(5)

print("📍 Most Frequent Matchups:")
for (home, away), count in top_matchups.items():
    print(f"   • {home} vs {away}: {count} matches")

# 5. Goal Difference Analysis
print("\n" + "🎯 " + "=" * 50)
print("🎯 GOAL DIFFERENCE & MATCH DOMINANCE")
print("🎯 " + "=" * 50)

df1819['Goal_Difference'] = abs(df1819['FTHG'] - df1819['FTAG'])
close_matches = (df1819['Goal_Difference'] <= 1).sum()
dominant_matches = (df1819['Goal_Difference'] >= 3).sum()

print(f"📍 Match Competitiveness:")
print(f"   • Close matches (≤1 goal difference): {close_matches} ({close_matches/len(df1819)*100:.1f}%)")
print(f"   • Dominant wins (≥3 goal difference): {dominant_matches} ({dominant_matches/len(df1819)*100:.1f}%)")

# 6. Visualization: Scatter Plot of Home vs Away Performance
print("\n" + "📈 " + "=" * 50)
print("📈 SCATTER PLOT: HOME VS AWAY PERFORMANCE")
print("📈 " + "=" * 50)

plt.figure(figsize=(12, 8))
plt.scatter(df_tot['home_goals'], df_tot['away_goals'], s=100, alpha=0.7, color='purple')
plt.xlabel('Home Goals', fontsize=12, fontweight='bold')
plt.ylabel('Away Goals', fontsize=12, fontweight='bold')
plt.title('Team Performance: Home Goals vs Away Goals', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)

# Add team labels
for i, team in enumerate(df_tot['team']):
    plt.annotate(team, (df_tot['home_goals'].iloc[i] + 0.1, df_tot['away_goals'].iloc[i] + 0.1),
                 fontsize=8, alpha=0.8)

plt.tight_layout()
plt.show()

# 7. Advanced Statistics
print("\n" + "🧮 " + "=" * 50)
print("🧮 ADVANCED STATISTICS")
print("🧮 " + "=" * 50)

# Calculate consistency metrics
home_consistency = df1819.groupby('HomeTeam')['FTHG'].std().mean()
away_consistency = df1819.groupby('AwayTeam')['FTAG'].std().mean()

print(f"📍 Performance Consistency:")
print(f"   • Home goal std dev (avg): {home_consistency:.2f}")
print(f"   • Away goal std dev (avg): {away_consistency:.2f}")

# 8. Predictive Insights
print("\n" + "🔮 " + "=" * 50)
print("🔮 PREDICTIVE INSIGHTS")
print("🔮 " + "=" * 50)

avg_goals_per_match = (df1819['FTHG'].sum() + df1819['FTAG'].sum()) / len(df1819)
home_win_rate = (df1819['FTHG'] > df1819['FTAG']).sum() / len(df1819) * 100
draw_rate = (df1819['FTHG'] == df1819['FTAG']).sum() / len(df1819) * 100

print(f"📍 Match Outcome Probabilities:")
print(f"   • Home wins: {home_win_rate:.1f}% of matches")
print(f"   • Draws: {draw_rate:.1f}% of matches")
print(f"   • Average goals per match: {avg_goals_per_match:.2f}")

print("\n" + "💭 " + "=" * 50)
print("💭 FEEDBACK & REFLECTION")
print("💭 " + "=" * 50)

feedback = {
    "assignment_quality": "Excellent - Comprehensive and practical",
    "learning_experience": "Great balance of guidance and exploration",
    "data_quality": "Good realistic football data",
    "skill_development": "Pandas, matplotlib, data analysis",
    "time_spent": "Approximately 2-3 hours",
    "favorite_part": "Goal analysis and visualization creation",
    "suggestions": "More real-world football analytics context"
}

print("📍 PERSONAL FEEDBACK:")
for key, value in feedback.items():
    print(f"   • {key.replace('_', ' ').title()}: {value}")

print("\n" + "🚀 " + "=" * 50)
print("🚀 FUTURE EXPLORATION IDEAS")
print("🚀 " + "=" * 50)

future_ideas = [
    "Player-level performance analysis",
    "Time-series analysis of team form",
    "Head-to-head rivalry deep dives",
    "Betting odds vs actual performance",
    "Injury impact on team performance",
    "Manager change effects",
    "Home advantage across different leagues"
]

for i, idea in enumerate(future_ideas, 1):
    print(f"   {i}. {idea}")

print(f"\n🎉 Thank you for this engaging data analysis assignment!")
print(f"📊 Total matches analyzed: {len(df1819):,}")
print(f"🏆 Teams analyzed: {len(df_tot)}")
print(f"⚽ Total goals scored: {df_tot['total'].sum():,}")
```
